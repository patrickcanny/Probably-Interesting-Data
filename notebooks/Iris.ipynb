A good candidate for clustering!

```python
import importlib.util
spec = importlib.util.spec_from_file_location("GMM", "../code/GMM.py")
GMM = importlib.util.module_from_spec(spec)
spec.loader.exec_module(GMM)
```

```python
i = iris.drop('Species', axis = 1)
iris_test = i[['SepalWidthCm', 'PetalLengthCm']].values
```

```python
GMM_mod = GMM.GMM(iris_test, 3, 75)
GMM_mod.fit()
```
